```python
import pandas as pd
import os
data_dir = 'data'
data_file_name = 'train_preprocess.csv'
data = pd.read_csv(os.path.join(data_dir, data_file_name), low_memory=True)
data
```

### 看讀進來ㄉ資料內容有哪些

```python
for ft in data.columns:
    print(ft," : ", data[ft].unique()," : ",len(data[ft].unique()))
```

### 算每個客戶（row）買了多少產品

```python
target_cols = ['ind_ahor_fin_ult1','ind_aval_fin_ult1','ind_cco_fin_ult1','ind_cder_fin_ult1','ind_cno_fin_ult1','ind_ctju_fin_ult1','ind_ctma_fin_ult1','ind_ctop_fin_ult1','ind_ctpp_fin_ult1','ind_deco_fin_ult1','ind_deme_fin_ult1','ind_dela_fin_ult1','ind_ecue_fin_ult1','ind_fond_fin_ult1','ind_hip_fin_ult1','ind_plan_fin_ult1','ind_pres_fin_ult1','ind_reca_fin_ult1','ind_tjcr_fin_ult1','ind_valo_fin_ult1','ind_viv_fin_ult1','ind_nomina_ult1','ind_nom_pens_ult1','ind_recibo_ult1']
target = data[target_cols]
target_cnt = target.sum(axis=1)
```

```python
print(target_cnt.unique(),len(target_cnt.unique()))
```

```python
np_target_cnt = target_cnt.to_numpy()
print(np_target_cnt.shape)
cnt = dict()
for value in np_target_cnt:
    if value not in cnt:
        cnt[value] = 1
    else:
        cnt[value] += 1
print(cnt)
```

### 把沒買產品的客戶剔除

```python
data_target_cnt = data.join(pd.DataFrame(target_cnt, columns = ['cnt']))
data_target_cnt
```

```python
filter_data = data_target_cnt[data_target_cnt['cnt'] !=0 ]
filter_data
```

### 找age 指定範圍的客戶

```python
filter_age_data = filter_data[(filter_data['age'] >= 30) & (filter_data['age'] < 40)]
filter_target = filter_age_data[target_cols]
filter_target
```

```python
import time
start = time.time()
association_data_list = []
for index, row in filter_target.iterrows():
    association_data = []
    for col_name in target_cols:
        if row[col_name] == 1:
            association_data.append(col_name)
    association_data_list.append(association_data)
end = time.time()
print(str(end-start)+'  seconds')
```

### 套Apriori

```python
from apyori import apriori

association_rules = apriori(association_data_list, min_support=0.05, min_confidence=0.2, min_lift=3, max_length=2) 
association_results = list(association_rules)
print(len(association_results))
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
